```python
# ---- Dark Theme Minimal Line Charts Collection ----
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
import numpy as np
from matplotlib.collections import LineCollection

# Chart 1: Neon Glow Lines
fig1, ax1 = plt.subplots(figsize=(14, 8))
ax1.set_facecolor('#0a0a0a')
fig1.patch.set_facecolor('#0a0a0a')

x = np.linspace(0, 12, 150)
y1 = 3.5 + np.sin(x * 0.7) * 1.8
y2 = 3.5 + np.cos(x * 0.5) * 1.5

# Glow effect with multiple lines
for lw, alpha in [(12, 0.1), (8, 0.15), (4, 0.3), (2, 1)]:
    ax1.plot(x, y1, color='#ff0080', linewidth=lw, alpha=alpha, solid_capstyle='round')
    ax1.plot(x, y2, color='#00ffff', linewidth=lw, alpha=alpha, solid_capstyle='round')

ax1.text(6, 6.5, 'Corin', fontsize=54, fontweight='bold', color='#ff0080',
         ha='center', va='center', family='sans-serif',
         path_effects=[pe.withStroke(linewidth=2, foreground='#0a0a0a')])
ax1.text(6, 0.8, 'Faife', fontsize=54, fontweight='bold', color='#00ffff',
         ha='center', va='center', family='sans-serif',
         path_effects=[pe.withStroke(linewidth=2, foreground='#0a0a0a')])

ax1.set_xlim(0, 12)
ax1.set_ylim(0, 7.5)
ax1.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Neon Glow Variation 4: Pulse Wave Interference
fig, ax = plt.subplots(figsize=(14, 8))
ax.set_facecolor('#000814')
fig.patch.set_facecolor('#000814')

x = np.linspace(0, 14, 350)

# Create interference pattern
neon_colors = ['#e63946', '#f77f00', '#fcbf49', '#eae2b7', '#90e0ef', '#00b4d8', '#0077b6', '#023e8a']

for i, color in enumerate(neon_colors):
    # Interference pattern with beat frequencies
    freq1 = 0.6 + i * 0.05
    freq2 = 0.65 + i * 0.05
    y = 4 + np.sin(x * freq1) * np.cos(x * 0.1 * (i + 1)) + np.sin(x * freq2 + i) * 0.8
    y = y + (i - 3.5) * 0.15
    
    # Glow effect
    for lw, alpha in [(20, 0.04), (14, 0.07), (8, 0.14), (3, 0.92)]:
        ax.plot(x, y, color=color, linewidth=lw, alpha=alpha, solid_capstyle='round')

ax.text(7, 7.3, 'Corin', fontsize=54, fontweight='bold', color='#e63946',
        ha='center', va='center', family='sans-serif',
        path_effects=[pe.withStroke(linewidth=3, foreground='#000814')])
ax.text(7, 0.7, 'Faife', fontsize=54, fontweight='bold', color='#00b4d8',
        ha='center', va='center', family='sans-serif',
        path_effects=[pe.withStroke(linewidth=3, foreground='#000814')])

ax.set_xlim(0, 14)
ax.set_ylim(0, 8)
ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Warm Sunset Palette - Legend Style 1
fig, ax = plt.subplots(figsize=(14, 8))
ax.set_facecolor('#0f0a1a')
fig.patch.set_facecolor('#0f0a1a')

x = np.linspace(0, 14, 350)

# Warm sunset gradient colors
warm_colors = ['#ff0a54', '#ff477e', "#272526", '#ff7096', '#ff85a1', '#ff99ac', '#fbb1bd', '#f9bec7', '#f7cad0', '#fae0e4']

for i, color in enumerate(warm_colors):
    phase = i * 0.55
    freq = 0.38 + i * 0.035
    amp = 1.5 - abs(i - 5) * 0.08
    y = 4 + amp * np.sin(x * freq + phase) * (1 + 0.25 * np.cos(x * 0.1))
    
    for lw, alpha in [(22, 0.035), (15, 0.06), (9, 0.12), (3.5, 0.92)]:
        ax.plot(x, y, color=color, linewidth=lw, alpha=alpha, solid_capstyle='round')

# Legend-style text box
legend_box = plt.Rectangle((0.4, 6.8), 4.2, 1.1, facecolor='#0f0a1a', edgecolor='#ff477e', 
                            linewidth=2, alpha=0.9, transform=ax.transData)
ax.add_patch(legend_box)

# Legend line samples
ax.plot([0.6, 1.4], [7.5, 7.5], color='#ff0a54', linewidth=3, solid_capstyle='round')
ax.plot([0.6, 1.4], [7.1, 7.1], color='#ff85a1', linewidth=3, solid_capstyle='round')

ax.text(1.6, 7.5, 'Corin', fontsize=22, fontweight='bold', color='#ff0a54',
        ha='left', va='center', family='sans-serif')
ax.text(1.6, 7.1, 'Faife', fontsize=22, fontweight='bold', color='#ff85a1',
        ha='left', va='center', family='sans-serif')

ax.set_xlim(0, 14)
ax.set_ylim(0, 8.2)
ax.axis('off')
plt.tight_layout()
plt.show()
```
